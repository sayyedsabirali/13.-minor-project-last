```python
import re
import string
import numpy as np
from difflib import get_close_matches
import spacy

class SimpleSymptomExtractor:
    def __init__(self, symptom_dict, threshold=0.75):
        self.symptom_dict = symptom_dict
        self.threshold = threshold
        self.symptom_terms = list(symptom_dict.keys())
        self.variation_to_symptom = {}
        for symptom, variations in self.symptom_dict.items():
            for variation in variations:
                self.variation_to_symptom[variation.lower()] = symptom
        for symptom in self.symptom_terms:
            normalized = symptom.replace('_', ' ').lower()
            self.variation_to_symptom[normalized] = symptom
        self.all_variations = list(self.variation_to_symptom.keys())
        try:
            self.nlp = spacy.load("en_core_web_md")
        except:
            print("Note: Using smaller spaCy model. For better results, install en_core_web_md")
            try:
                self.nlp = spacy.load("en_core_web_sm")
            except:
                print("Error: spaCy model not loaded. Run: python -m spacy download en_core_web_sm")
                self.nlp = None
        self.severity_modifiers = {
            'severe': 1.0, 'extreme': 1.0, 'serious': 0.9, 'significant': 0.9,
            'moderate': 0.7, 'mild': 0.5, 'slight': 0.4, 'minor': 0.3
        }
        self.frequency_modifiers = {
            'constant': 1.0, 'always': 1.0, 'persistent': 0.9, 'frequent': 0.8,
            'often': 0.7, 'sometimes': 0.5, 'occasional': 0.4, 'rarely': 0.3
        }
    def preprocess_text(self, text):
        text = text.lower()
        if self.nlp:
            doc = self.nlp(text)
            clean_tokens = [token.text for token in doc if not token.is_stop and not token.is_punct]
            clean_text = ' '.join(clean_tokens)
            return clean_text, doc
        else:
            text = text.translate(str.maketrans('', '', string.punctuation))
            return text, None

    def find_direct_matches(self, text):
        matches = {}
        for variation, symptom in self.variation_to_symptom.items():
            if variation in text and symptom not in matches:
                matches[symptom] = 1.0  # Full confidence for direct matches
        return matches
    def find_fuzzy_matches(self, text):
        matches = {}
        words = text.split()
        for word in words:
            if len(word) > 3:
                close_matches = get_close_matches(word, self.all_variations, n=1, cutoff=0.8)
                for match in close_matches:
                    symptom = self.variation_to_symptom.get(match)
                    if symptom and symptom not in matches:
                        matches[symptom] = 0.8
        if len(words) >= 2:
            for i in range(len(words) - 1):
                bigram = words[i] + ' ' + words[i+1]
                close_matches = get_close_matches(bigram, self.all_variations, n=1, cutoff=0.75)
                for match in close_matches:
                    symptom = self.variation_to_symptom.get(match)
                    if symptom and symptom not in matches:
                        matches[symptom] = 0.85
        if len(words) >= 3:
            for i in range(len(words) - 2):
                trigram = words[i] + ' ' + words[i+1] + ' ' + words[i+2]
                close_matches = get_close_matches(trigram, self.all_variations, n=1, cutoff=0.7)
                for match in close_matches:
                    symptom = self.variation_to_symptom.get(match)
                    if symptom and symptom not in matches:
                        matches[symptom] = 0.9

        return matches

    def find_semantic_matches(self, doc):
        if not self.nlp or not doc:
            return {}
        matches = {}
        variation_docs = {var: self.nlp(var) for var in self.all_variations}
        for sent in doc.sents:
            for variation, var_doc in variation_docs.items():
                similarity = sent.similarity(var_doc)
                if similarity >= self.threshold:
                    symptom = self.variation_to_symptom.get(variation)
                    if symptom and (symptom not in matches or similarity > matches[symptom]):
                        matches[symptom] = float(similarity)
        return matches

    def check_negations(self, doc, symptom_matches):
        if not self.nlp or not doc:
            return symptom_matches
        filtered_matches = symptom_matches.copy()
        for symptom in list(filtered_matches.keys()):
            normalized = symptom.replace('_', ' ').lower()
            for sent in doc.sents:
                sent_text = sent.text.lower()
                if normalized not in sent_text:
                    continue
                negation_words = ["no", "not", "don't", "doesn't", "didn't", "haven't",
                                 "hasn't", "won't", "wouldn't", "can't", "cannot"]
                for neg_word in negation_words:
                    pattern = f"{neg_word}\\s+\\w+(?:\\s+\\w+){{0,3}}\\s+{re.escape(normalized)}"
                    if re.search(pattern, sent_text):
                        if symptom in filtered_matches:
                            del filtered_matches[symptom]
                        break

        return filtered_matches
    def detect_modifiers(self, doc, symptoms):
        severity_info = {}
        frequency_info = {}

        if not self.nlp or not doc:
            for symptom in symptoms:
                severity_info[symptom] = 'unspecified'
                frequency_info[symptom] = 'unspecified'
            return severity_info, frequency_info
        for symptom in symptoms:
            normalized = symptom.replace('_', ' ').lower()
            severity_found = False
            frequency_found = False
            for sent in doc.sents:
                sent_text = sent.text.lower()
                if normalized not in sent_text:
                    continue
                for modifier, _ in self.severity_modifiers.items():
                    if modifier in sent_text and not severity_found:
                        severity_info[symptom] = modifier
                        severity_found = True
                        break
                for modifier, _ in self.frequency_modifiers.items():
                    if modifier in sent_text and not frequency_found:
                        frequency_info[symptom] = modifier
                        frequency_found = True
                        break
            if not severity_found:
                severity_info[symptom] = 'unspecified'
            if not frequency_found:
                frequency_info[symptom] = 'unspecified'

        return severity_info, frequency_info

    def extract_symptoms(self, query):
        clean_text, doc = self.preprocess_text(query)
        direct_matches = self.find_direct_matches(clean_text)
        fuzzy_matches = self.find_fuzzy_matches(clean_text)
        semantic_matches = self.find_semantic_matches(doc) if doc else {}
        all_matches = {}
        for matches in [direct_matches, fuzzy_matches, semantic_matches]:
            for symptom, confidence in matches.items():
                if symptom not in all_matches or confidence > all_matches[symptom]:
                    all_matches[symptom] = confidence
        filtered_matches = self.check_negations(doc, all_matches)
        severity, frequency = self.detect_modifiers(doc, filtered_matches)
        sorted_symptoms = sorted(filtered_matches.items(), key=lambda x: x[1], reverse=True)
        final_symptoms = {symptom: conf for symptom, conf in sorted_symptoms if conf >= self.threshold}
        results = {
            'symptoms': list(final_symptoms.keys()),
            'confidence': final_symptoms,
            'severity': {s: severity.get(s, 'unspecified') for s in final_symptoms},
            'frequency': {s: frequency.get(s, 'unspecified') for s in final_symptoms}
        }
        return results


# Example usage:
if __name__ == "__main__":
    extractor = SimpleSymptomExtractor(symptom_dict)
    query ="I've been having chest pain and shortness of breath, especially when I exercise. Sometimes I feel my heart racing."
    results = extractor.extract_symptoms(query)
    print("Detected symptoms:")
    for symptom in results['symptoms']:
        confidence = results['confidence'][symptom]
        severity = results['severity'][symptom]
        frequency = results['frequency'][symptom]

        print(f"- {symptom.replace('_', ' ')} (Confidence: {confidence:.2f}, Severity: {severity}, Frequency: {frequency})")
```

```python
import pandas as pd
import torch
import numpy as np
from transformers import AutoTokenizer, AutoModel
from sklearn.model_selection import train_test_split
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_recall_fscore_support
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import time
import os
import json

SEED = 42
MAX_LENGTH = 256
BATCH_SIZE = 16
LEARNING_RATE = 2e-5
EPOCHS = 10 
MODEL_NAME = "microsoft/BiomedNLP-PubMedBERT-base-uncased-abstract-fulltext"
torch.manual_seed(SEED)
np.random.seed(SEED)
torch.backends.cudnn.deterministic = True
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)
class EnhancedSymptomClassifier(torch.nn.Module):
    def __init__(self, base_model_name, num_classes, hidden_size=768, dropout_rate=0.3):
        super(EnhancedSymptomClassifier, self).__init__()
        self.bert = AutoModel.from_pretrained(base_model_name)
        self.attention = torch.nn.Sequential(
            torch.nn.Linear(hidden_size, 128),
            torch.nn.Tanh(),
            torch.nn.Linear(128, 1),
            torch.nn.Softmax(dim=1)
        )
        self.classifier = torch.nn.Sequential(
            torch.nn.Linear(hidden_size, 512),
            torch.nn.LeakyReLU(),
            torch.nn.BatchNorm1d(512),
            torch.nn.Dropout(dropout_rate),
            torch.nn.Linear(512, 256),
            torch.nn.LeakyReLU(),
            torch.nn.BatchNorm1d(256),
            torch.nn.Dropout(dropout_rate),
            torch.nn.Linear(256, 128),
            torch.nn.LeakyReLU(),
            torch.nn.BatchNorm1d(128),
            torch.nn.Dropout(dropout_rate),
            torch.nn.Linear(128, num_classes)
        )
        self.symptom_classifier = torch.nn.Sequential(
            torch.nn.Linear(hidden_size, 128),
            torch.nn.LeakyReLU(),
            torch.nn.Dropout(dropout_rate),
            torch.nn.Linear(128, 64),
            torch.nn.LeakyReLU(),
            torch.nn.Linear(64, num_classes)
        )

    def forward(self, input_ids, attention_mask):
        outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        sequence_output = outputs.last_hidden_state
        attention_weights = self.attention(sequence_output).transpose(-1, -2)
        weighted_output = torch.matmul(attention_weights, sequence_output).squeeze(1)
        main_logits = self.classifier(weighted_output)
        symptom_logits = self.symptom_classifier(outputs.last_hidden_state[:, 0, :])
        logits = main_logits * 0.7 + symptom_logits * 0.3
        return logits
def load_preprocess_data(file_path):
    df = pd.read_csv('dataset.csv')
    symptom_cols = [col for col in df.columns if col.startswith('Symptom_')]
    for col in symptom_cols:
        df[col] = df[col].fillna('').astype(str)
    df["symptoms"] = df[symptom_cols].apply(
        lambda x: " ".join([
            f"patient presents with {str(i).lower().replace('_', ' ')}"
            for i in x if str(i).lower() != 'nan' and str(i).strip()
        ]),
        axis=1
    )
    df["symptoms"] = "Medical case: " + df["symptoms"]
    disease_map = {disease: idx for idx, disease in enumerate(df["Disease"].unique())}
    df["disease_label"] = df["Disease"].map(disease_map)
    processed_df = df[["symptoms", "Disease", "disease_label"]]

    return processed_df, disease_map

# Improved training with learning rate scheduling
def train_model(model, train_dataloader, val_dataloader, optimizer, criterion, device, output_dir="models", epochs=EPOCHS):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    model_save_path = os.path.join(output_dir, "best_enhanced_pubmedbert_model.pt")
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, mode='max', factor=0.5, patience=2, verbose=True
    )
    best_accuracy = 0
    history = {
        'train_loss': [],
        'val_loss': [],
        'val_accuracy': [],
        'val_f1': []
    }
    for epoch in range(epochs):
        start_time = time.time()
        model.train()
        total_loss = 0
        progress_bar = tqdm(train_dataloader, desc=f"Epoch {epoch+1}/{epochs}")
        for batch in progress_bar:
            optimizer.zero_grad()
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)
            labels = batch["labels"].to(device)
            logits = model(input_ids, attention_mask)
            loss = criterion(logits, labels)
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()
            total_loss += loss.item()
            progress_bar.set_postfix({'loss': loss.item()})

        avg_train_loss = total_loss / len(train_dataloader)
        history['train_loss'].append(avg_train_loss)

        # Validation phase
        val_metrics = evaluate_model(model, val_dataloader, criterion, device, compute_metrics=True)
        history['val_loss'].append(val_metrics['loss'])
        history['val_accuracy'].append(val_metrics['accuracy'])
        history['val_f1'].append(val_metrics['f1'])
        scheduler.step(val_metrics['accuracy'])
        epoch_time = time.time() - start_time
        print(f"Epoch {epoch+1}/{epochs} - Time: {epoch_time:.1f}s - Train Loss: {avg_train_loss:.4f} - "
              f"Val Loss: {val_metrics['loss']:.4f} - Val Accuracy: {val_metrics['accuracy']:.4f} - "
              f"Val F1: {val_metrics['f1']:.4f}")
        if val_metrics['accuracy'] > best_accuracy:
            best_accuracy = val_metrics['accuracy']
            torch.save(model.state_dict(), model_save_path)
            print(f"New best model saved at {model_save_path}")
    history_path = os.path.join(output_dir, "enhanced_pubmedbert_training_history.json")
    with open(history_path, 'w') as f:
        json.dump(history, f)
    return history
class SymptomDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_len=MAX_LENGTH):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len
    def __len__(self):
        return len(self.texts)
    def __getitem__(self, idx):
        encoding = self.tokenizer(
            self.texts[idx],
            max_length=self.max_len,
            padding='max_length',
            truncation=True,
            return_tensors="pt"
        )
        return {
            "input_ids": encoding["input_ids"].squeeze(0),
            "attention_mask": encoding["attention_mask"].squeeze(0),
            "labels": torch.tensor(self.labels[idx], dtype=torch.long),
        }
def evaluate_model(model, dataloader, criterion, device, compute_metrics=False):
    model.eval()
    predictions = []
    true_labels = []
    total_loss = 0
    with torch.no_grad():
        for batch in dataloader:
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)
            labels = batch["labels"].to(device)
            # Forward pass
            logits = model(input_ids, attention_mask)
            loss = criterion(logits, labels)
            total_loss += loss.item()
            preds = torch.argmax(logits, dim=1).cpu().numpy()
            labels = labels.cpu().numpy()
            predictions.extend(preds)
            true_labels.extend(labels)
    avg_loss = total_loss / len(dataloader)
    accuracy = accuracy_score(true_labels, predictions)
    metrics = {
        'loss': avg_loss,
        'accuracy': accuracy
    }
    if compute_metrics:
        # Calculate precision, recall, F1 (weighted by support)
        precision, recall, f1, _ = precision_recall_fscore_support(
            true_labels, predictions, average='weighted'
        )
        metrics.update({
            'precision': precision,
            'recall': recall,
            'f1': f1,
            'predictions': predictions,
            'true_labels': true_labels
        })
    return metrics
def predict_disease(model, tokenizer, extracted_symptoms_dict, device, disease_map, disease_symptom_mapping=None):
    normalized_symptoms = {symptom: score/100 for symptom, score in extracted_symptoms_dict.items()}
    sorted_symptoms = sorted(normalized_symptoms.items(), key=lambda x: x[1], reverse=True)
    queries = []
    symptom_texts = []
    for symptom, confidence in sorted_symptoms:
        symptom_text = symptom.replace('_', ' ')
        if confidence > 0.8:
            symptom_texts.extend([symptom_text] * 3) 
        elif confidence > 0.6:
            symptom_texts.extend([symptom_text] * 2)  
        elif confidence > 0.3:
            symptom_texts.append(symptom_text) 

    if symptom_texts:
        queries.append("Patient symptoms: " + ", ".join(symptom_texts))
    case_description = ["Medical case:"]
    for symptom, confidence in sorted_symptoms:
        if confidence > 0.3:  
            symptom_text = symptom.replace('_', ' ')
            if confidence > 0.8:
                descriptor = "severe"
            elif confidence > 0.6:
                descriptor = "moderate"
            else:
                descriptor = "mild"
            case_description.append(f"Patient presents with {descriptor} {symptom_text}.")

    if len(case_description) > 1:
        queries.append(" ".join(case_description))
    high_conf_symptoms = [s for s, c in sorted_symptoms if c > 0.7]
    if high_conf_symptoms:
        formatted_symptoms = [s.replace('_', ' ') for s in high_conf_symptoms]
        queries.append(f"Primary symptoms: {', '.join(formatted_symptoms)}. What disease matches these symptoms?")
    if not queries:
        top_symptoms = [s.replace('_', ' ') for s, _ in sorted_symptoms[:3]]
        queries.append("Patient symptoms: " + ", ".join(top_symptoms))
    all_probabilities = []
    for query in queries:
        # Tokenize
        inputs = tokenizer(
            query,
            max_length=MAX_LENGTH,
            padding='max_length',
            truncation=True,
            return_tensors="pt"
        ).to(device)
        model.eval()
        with torch.no_grad():
            logits = model(inputs['input_ids'], inputs['attention_mask'])
            probabilities = torch.nn.functional.softmax(logits, dim=1)[0]
            all_probabilities.append(probabilities)
    avg_probabilities = torch.mean(torch.stack(all_probabilities), dim=0)
    if disease_symptom_mapping:
        correlation_scores = torch.zeros(len(disease_map)).to(device)
        for disease_idx, disease in enumerate(disease_map.keys()):
            if disease in disease_symptom_mapping:
                common_symptoms = disease_symptom_mapping[disease]
                matched_symptoms = set(extracted_symptoms_dict.keys()).intersection(set(common_symptoms))
                score = sum(extracted_symptoms_dict[symptom]/100 for symptom in matched_symptoms)
                correlation_scores[disease_idx] = score
        if torch.sum(correlation_scores) > 0:
            correlation_scores = correlation_scores / torch.sum(correlation_scores)
            avg_probabilities = avg_probabilities * 0.7 + correlation_scores * 0.3
    top_k = 3
    topk_probs, topk_indices = torch.topk(avg_probabilities, top_k)
    reverse_disease_map = {idx: disease for disease, idx in disease_map.items()}

    results = []
    for i in range(top_k):
        disease_idx = topk_indices[i].item()
        disease_name = reverse_disease_map[disease_idx]
        probability = topk_probs[i].item()
        results.append({"disease": disease_name, "probability": probability})
    return results

def create_symptom_disease_mapping(df):
    disease_symptom_map = {}
    for disease in df["Disease"].unique():
        disease_rows = df[df["Disease"] == disease]
        all_symptoms = []
        for _, row in disease_rows.iterrows():
            symptom_cols = [col for col in df.columns if col.startswith('Symptom_')]
            symptoms = [str(row[col]) for col in symptom_cols
                       if str(row[col]).lower() != 'nan' and str(row[col]).strip()]
            all_symptoms.extend(symptoms)
        from collections import Counter
        symptom_counts = Counter(all_symptoms)
        threshold = len(disease_rows) * 0.25
        common_symptoms = [symptom for symptom, count in symptom_counts.items()
                         if count >= threshold]

        disease_symptom_map[disease] = common_symptoms

    return disease_symptom_map
def main():
    # Set device
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")
    print("Loading and preprocessing data...")
    file_path = "/content/symptoms.csv" 
    df, disease_map = load_preprocess_data(file_path)
    print(f"Dataset loaded with {len(df)} rows and {len(disease_map)} diseases")
    disease_symptom_mapping = create_symptom_disease_mapping(df)
    train_texts, val_texts, train_labels, val_labels = train_test_split(
        df["symptoms"].tolist(), df["disease_label"].tolist(), test_size=0.2, random_state=SEED
    )
    tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
    train_dataset = SymptomDataset(train_texts, train_labels, tokenizer)
    val_dataset = SymptomDataset(val_texts, val_labels, tokenizer)
    train_dataloader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
    val_dataloader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)
    model = EnhancedSymptomClassifier(MODEL_NAME, num_classes=len(disease_map))
    model.to(device)
    optimizer = torch.optim.AdamW(
        [
            {'params': model.bert.parameters(), 'lr': LEARNING_RATE / 10, 'weight_decay': 0.01},
            {'params': model.classifier.parameters(), 'lr': LEARNING_RATE, 'weight_decay': 0.01},
            {'params': model.symptom_classifier.parameters(), 'lr': LEARNING_RATE, 'weight_decay': 0.01},
            {'params': model.attention.parameters(), 'lr': LEARNING_RATE, 'weight_decay': 0.01}
        ]
    )
    class_counts = np.bincount(train_labels)
    class_weights = 1. / torch.tensor(class_counts, dtype=torch.float)
    class_weights = class_weights / class_weights.sum() * len(class_counts)
    class_weights = class_weights.to(device)
    criterion = torch.nn.CrossEntropyLoss(weight=class_weights)
    model_save_path = "models/best_enhanced_pubmedbert_model.pt"
    if os.path.exists(model_save_path):
        print(f"\nLoading pre-trained model from: {model_save_path}")
        model.load_state_dict(torch.load(model_save_path, map_location=device))
    else:
        print("\nTraining enhanced model...")
        history = train_model(
            model,
            train_dataloader,
            val_dataloader,
            optimizer,
            criterion,
            device,
            epochs=EPOCHS
        )

    # Example of using the improved prediction function
    test_symptoms = extracted_symptoms

    results = predict_disease(
        model,
        tokenizer,
        test_symptoms,
        device,
        disease_map,
        disease_symptom_mapping  # Include the disease-symptom mapping
    )

    print("\nTop 3 predicted diseases:")
    for result in results:
        print(f"{result['disease']}: {result['probability']:.4f}")

    # Save the improved model for production
    torch.save({
        'model_state_dict': model.state_dict(),
        'disease_map': disease_map,
        'disease_symptom_mapping': disease_symptom_mapping,
        'model_config': {
            'model_name': MODEL_NAME,
            'num_classes': len(disease_map)
        }
    }, "enhanced_production_model.pth")

    print("\nEnhanced model saved for production use at enhanced_production_model.pth")

if __name__ == "__main__":
    main()

#
```
